```python
import numpy as np
import matplotlib.pyplot as plt

data_nod = np.loadtxt("fig2_nod.dat").reshape(80001,2,21,5)
data_sd  = np.loadtxt('fig2_sd.dat').reshape(80001,2,21,5)
```

```python
fig_set = {
    "axes.linewidth": .5,
    "axes.labelsize": 10,
    "axes.labelpad": 4.0,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8,
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "xtick.minor.size": 2,
    "xtick.minor.width" : 0.5,
    "ytick.minor.size": 3,
    "ytick.minor.width" : 0.5,
    "xtick.minor.visible" : True,
    "ytick.minor.visible" : True,
    "xtick.top" : True,
    "ytick.right" : True,
    "font.size": 10,
    "font.family": 'sans-serif',
    "legend.frameon" : False,
    "xtick.direction": 'in',  
    "ytick.direction": 'in' 
}

plt.rcParams.update(fig_set)
```

```python
z3s = data_sd[:,0,0,2]
h3s = data_sd[:,0,:,3]
z3_nod = data_nod[:,0,0,2]
h3_nod = data_nod[:,0,:,3]
```

```python
dpi = 200
fig, ax= plt.subplots(figsize=(4,3), dpi=dpi)
plt.yscale("log")
ax.set_xlabel("$\\xi^3$, $\mu$")
ax.set_ylabel("Metrics along the field line \n $\sqrt{g_{33}},\,h_{\mu}$")
ax.plot(z3_nod[:],h3_nod[:,10], label=("$\sqrt{g_{33}}$")  , linewidth=.8)
ax.plot(z3s[:],h3s[:,10], label=("$h_{\mu}$")  , linewidth=.8)
plt.legend()
plt.savefig("fig2.png", bbox_inches='tight' )
```
