```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("omg_kperp.d")

print(data.shape)
```

```python
#d = data.reshape(200,5,6)
d1 = data[:399,:]
d2 = data[400:799,:]
d3 = data[800:1199,:]
d4 = data[1200:1599,:]
d5 = data[1600:1999,:]
```

```python
plt.xlabel("$k_x$")
plt.ylabel("Re[$\Omega$]")

plt.plot(d1[:,2],d1[:,4])
plt.plot(d2[:,2],d2[:,4], linestyle = 'dotted')
plt.plot(d3[:,2],d3[:,4], linestyle = 'dotted')
plt.plot(d4[:,2],d4[:,4], linestyle = 'dotted')
plt.plot(d5[:,2],d5[:,4], linestyle = 'dotted')
```

```python
plt.xlabel("$k_x$")
plt.ylabel("Im[$\Omega$]")
plt.xlim(0,20000)
plt.plot(d1[:,2],d1[:,5])
plt.plot(d2[:,2],d2[:,5], linestyle = 'dotted')
plt.plot(d3[:,2],d3[:,5], linestyle = 'dotted')
plt.plot(d4[:,2],d4[:,5], linestyle = 'dotted')
plt.plot(d5[:,2],d5[:,5], linestyle = 'dotted')
```

```python
#基本波について
kp = d1[:,2]
Ro = d1[:,4]
Io = d1[:,5]

plt.xlim(0,4000)
plt.plot(kp,Ro)
plt.show()
plt.xlim(0,4000)
plt.plot(kp,Io)
plt.show()
```

```python
i = 0
for i in range(len(kp)) :
    if Io[i] > Io.max() * 0.9999999 :
        print('kp=',kp[i],'OMG=',Ro[i],Io[i])
        kpm = kp[i]
        Iom = Io[i]

plt.xlim(0,4000)
plt.plot(kp,Io)
plt.scatter(kpm,Iom , color='orange')
```

```python
kx = np.sqrt(1/5) * kpm
ky = 2 * kx

print(kx,ky , np.sqrt(kx**2 + ky**2) )
```
