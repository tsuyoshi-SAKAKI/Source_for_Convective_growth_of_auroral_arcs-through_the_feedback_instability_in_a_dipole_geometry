```python
import numpy as np
import matplotlib.pyplot as plt

#data = np.loadtxt("omg_k_l10_test.d")
data = np.loadtxt("l1079_lat2000master.d")

print(data.shape)
```

```python
fig_set = {
    "axes.linewidth": .5,
    "axes.labelsize": 10,
    "axes.labelpad": 4.0,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8,
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "xtick.minor.size": 2,
    "xtick.minor.width" : 0.5,
    "ytick.minor.size": 3,
    "ytick.minor.width" : 0.5,
    "xtick.minor.visible" : True,
    "ytick.minor.visible" : True,
    "xtick.top" : True,
    "ytick.right" : True,
    "font.size": 10,
##    "font.family": 'Times New Roman',
##    "mathtext.fontset": 'stix'
    "font.family": 'sans-serif',
#    "font.family": 'serif',
#    "font.weight" : "bold"
#    "legend.fancybox": True
#    "legend.framealpha": .5 # 凡例の透明度
    "legend.frameon" : False,
    "xtick.direction": 'in',  ## デフォルト: out
    "ytick.direction": 'in'  ## デフォルト: out
}

plt.rcParams.update(fig_set)
```

```python
nl = 499
nt = 5
jloop = nt
data = data.reshape(nt,nl,6)

kpa_r = data[:,:,0]
kpa_i = data[:,:,1]
kpe   = data[:,:,2]
theta = data[:,:,3]
omg_r = data[:,:,4]
omg_i = data[:,:,5]
```

```python
str = ["solid","dashed","dashdot","dotted","dotted"]
dpi = 250
fig, ax= plt.subplots(figsize=(6,4), dpi=dpi)
plt.xlabel("$k'_{\\xi^1}$")
plt.ylabel("$\mathrm{Re}[\Omega]$")
ax.set_xlim(0,20000)
for i in range(jloop):
    ax.plot(kpe[i,:],omg_r[i,:],linestyle=str[i])
```

```python
str = ["solid","dashed","dashdot","dotted","dotted"]
dpi = 250
fig, ax= plt.subplots(figsize=(6,4), dpi=dpi)
plt.xlabel("$k'_{\\xi^1}$")
plt.ylabel("$\mathrm{Im}[\Omega]$")
ax.set_xlim(0,20000)
for i in range(jloop):
    ax.plot(kpe[i,:],omg_i[i,:],linestyle=str[i])
```
