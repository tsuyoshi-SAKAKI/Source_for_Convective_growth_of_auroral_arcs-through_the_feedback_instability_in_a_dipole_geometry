```python
data=np.loadtxt("Fld/rmhdfs_r0.1.slab_linear.run07D4_j-phi.g0000.ion.001")
print(data.shape)
```

```python
### Reshape data -> x, dns, omg, cpj ###
dt=1.0
nx=1 + 2* 1280
nt=int(data.shape[0]/nx)
print(nt)
print(data.shape[0])

x=data[0:nx,0]
xx = 90-np.arcsin(np.sqrt(x))*180/np.pi

t=dt * np.arange(nt)
dns=(data[:,1]+1j*data[:,2]).reshape(nt,nx) #密度のdataを実部＋虚部に書き直し、それを縦軸nt=1601,横軸nx=2561でreshape

print(x.shape)
print(t.shape)
print(dns.shape)
print(dns.max())

dx=x[1]-x[0]
kx=fft.fftfreq(nx, dx/(2.0*np.pi)) # kx = 0, kx_min, 2*kx_min, ..., -2*kx_min, -kx_min
x2= fft.ifft(kx)

print(kx[0],kx[1])
print(x[0],x2[0])
print(90-xx.max(),90-xx.min())
```

```python
#先にxにフーリエ変換し、その後にstft
#x = 90-np.arcsin(np.sqrt(xx))*180/np.pi
x = data[0:nx,0]
# x方向にフーリエ変換。入力: dns(t,x) → 出力: dns_tk(t,kx)
dns_tk = fft.fft(dns, axis=-1) / nx # divide by nx for normalization
 # 波数軸 kx を定義。
dx=x[1]-x[0]
kx=fft.fftfreq(nx, dx/(2.0*np.pi)) # kx = 0, kx_min, 2*kx_min, ..., -2*kx_min, -kx_min
print(kx.shape)
print(x.shape)
print(dns_tk.shape)

fig = plt.figure()
ax = fig.add_subplot(111)

it=50
ax.plot(kx[0:int(nx/6)],np.abs(dns_tk[it,0:int(nx/6)]),label="t={}".format(t[it]))
it=100
ax.plot(kx[0:int(nx/6)],np.abs(dns_tk[it,0:int(nx/6)]),label="t={}".format(t[it]))
ax.set_xlabel(r"Wevenumber $k_{\xi^1}$")
ax.set_ylabel(r"$|n_k|$")
ax.legend()
plt.show()
twindow=800.0
dt=t[1]-t[0]               # Time step size for sampling
nwindow = int(twindow/dt)  # Window length of short-time Fourier transform
# t方向に短時間フーリエ変換(short-time Fourier transform) 入力: dns_tk(t,kx) , 出力: dns_wxt(omega,kx,time)
freq,time,dns_wxt = signal.stft(dns_tk[:,:], fs=1.0/dt, window=signal.get_window(('hann'),nwindow),
                                    nperseg=nwindow, noverlap=nwindow/2, return_onesided=False, boundary=None, axis=0)


plt.plot(signal.get_window('hann',nwindow) , linestyle="dotted")
plt.plot(signal.get_window(('tukey',1.0),nwindow))
plt.show()
print(dns_wxt.shape)
print(time.shape)
```

```python
omega = 2*np.pi*freq # 周波数軸 omega
print(freq.shape)
print(omega.shape)
print(time)
print(np.abs(dns_tk[it,0:int(nx/6)]).max())
```
