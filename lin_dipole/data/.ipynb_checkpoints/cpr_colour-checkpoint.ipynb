```python
import numpy as np
from scipy import signal, fft
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import scipy as sp
from scipy.optimize import curve_fit 
```

```python
data=np.loadtxt("./Fld/rmhdfs_r0.1.slab_linear.run07D4_j-phi.g0000.ion.001")
```

```python
fig_set = {
    "axes.linewidth": .5,
    "axes.labelsize": 10,
    "axes.labelpad": 4.0,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8,
    "xtick.major.size": 3,
    "ytick.major.size": 3,
    "ytick.right" : True,
    "xtick.minor.visible" : False,
    "ytick.minor.visible" : True,
    "font.size": 10,
    "font.family": 'sans-serif',
    "legend.framealpha": .5, # 凡例の透明度
    "legend.frameon" : False,
    "xtick.direction": 'in',  ## デフォルト: out
    "ytick.direction": 'in'  ## デフォルト: out
}
plt.rcParams.update(fig_set)
```

```python
dt=1.0
nx=1280*2+1
nt=int(data.shape[0]/nx)
x=data[0:nx,0]
xx = 90-np.arcsin(np.sqrt(x))*180/np.pi
xc = np.arcsin(np.sqrt(x))*180/np.pi
t=dt * np.arange(nt)
cpr=(data[:,5]+1j*data[:,6]).reshape(nt,nx) #密度のdataを実部＋虚部に書き直し、それを縦軸nt=1601,横軸nx=2561でreshape
cpr0 =  (np.abs( cpr[0,:].real ) ).max()
print(cpr0)
print( (np.abs( cpr[:,:].real ) ).max() / cpr0 )
print( np.abs(cpr.real).max() ,  np.abs(cpr.real).max()/cpr0 )
cpr = cpr.real/cpr0  # 初期振幅で規格化
```

```python
### 2D color plot n(t,x) ###
figsize_px = np.array([1280, 960])
dpi = 300
fig, ax= plt.subplots(figsize=(4,3), dpi=dpi)
# COLATITULE
quad = ax.pcolormesh(xc,t,cpr.real,rasterized=True,cmap="viridis")
cbar = fig.colorbar(quad)
cbar.ax.tick_params(labelsize=8,width =0.4)
cbar.outline.set_linewidth(0.4)
cbar.set_label("Current density")
ax.tick_params( labelsize= 8, width=0.5)
ax2 = ax.twiny()
ax2.tick_params( labelsize= 8, width=0.5)
ax2.set_xlabel("$\sin^2\\theta$" )
ax2.set_xlim(np.sin(19/180*np.pi)**2, np.sin(xc.max()/180*np.pi)**2 )
# COLATITUDE
ax.set_xlim(19, xc.max())
ax.set_ylim(0, 1700)
# COLATITUDE
ax.set_xlabel( "Colatitude $(^\mathrm{\circ})$") 
ax.set_ylabel("Time ($\mathrm{R_I/V_A}$)")
#plt.savefig("./Fig/fig_ion_cpr.pdf", bbox_inches='tight', pad_inches=0.1)
plt.show()
```

```python
a = np.zeros((len(cpr.real[:,0]), len(cpr.real[0,:])))
print(a.shape)

for i in range(nt):
    a[i,:] = np.abs(signal.hilbert(cpr.real[i,:]))

am = np.zeros((len(cpr.real[:,0])))
xm=  np.zeros((len(cpr.real[:,0])))
print(am.shape)
for m in range(nt):
    for n in range(nx):
        if a[m,n] == a[m,:].max():
            am[m] = a[m,n]
            xm[m] = xc[n]
```

```python
### 2D color plot n(t,x) ###
figsize_px = np.array([1280, 960])
dpi = 300
figsize_inch = figsize_px / dpi
print("figsize", figsize_inch)  # figsize (6.4, 4.8)
fig, ax= plt.subplots(figsize=(4,3), dpi=dpi)

#ax = fig.add_subplot(111)
ax.plot(xm,t,linewidth=.8, color="white", linestyle="solid" , alpha=1.0 ,rasterized=True)   ######. 群速度
# COLATITULE
quad = ax.pcolormesh(xc,t,cpr.real,rasterized=True,cmap="viridis")
cbar = fig.colorbar(quad)#, shrink=1.0, aspect=10.0)
cbar.set_label('Density perturbation', fontsize=10) # 文字サイズ14でラベルを付ける)
cbar.ax.tick_params(labelsize=8,width =0.4)
cbar.outline.set_linewidth(0.4)
cbar.set_label("Current density")

ax.tick_params( labelsize= 8, width=0.5)
ax2 = ax.twiny()
ax2.tick_params( labelsize= 8, width=0.5)
ax2.set_xlabel("$\sin^2\\theta$" )
ax2.set_xlim(np.sin(19/180*np.pi)**2, np.sin(xc.max()/180*np.pi)**2 )
# COLATITUDE
ax.set_xlim(19, xc.max())
ax.set_ylim(0, 1700)
# COLATITUDE
ax.set_xlabel( "Colatitude $(^\mathrm{\circ})$") 
ax.set_ylabel("Time ($\mathrm{R_I/V_A}$)")
#plt.savefig("fig3_vg.pdf", bbox_inches='tight', pad_inches=0.1)
plt.show()
```

```python
####### curve fit ##########
it = 300

data_xd = t
data_yd = xm/180*np.pi

parameter_initial = np.array([1 ,0.006])
def func(x, a, b ):
    fd = b*x + a
    return fd

poptd, pcovd = curve_fit(func,data_xd, data_yd, parameter_initial)

print('a*exp(b*t) -> a,b=', poptd, pcovd)

residualsd =  data_yd - func(data_xd, poptd[0],poptd[1]) #残差
rssd = np.sum(residualsd**2)  #residual sum of squares = rss
tssd = np.sum((data_yd-np.mean(data_yd))**2)  #total sum of squares = tss
r_squaredd = 1 - (rssd / tssd)

print(r_squaredd)
```

```python
plt.yscale("log")
plt.xlabel('$t~[L_x/V_A]$')
plt.ylabel('$log|J|$')
#ax = fig.add_subplot(111)
plt.plot(data_xd,data_yd , label="icr")
plt.plot(data_xd,func(data_xd, poptd[0],poptd[1]) , label='gr={:.6f}'.format(poptd[1]))
plt.legend()
plt.show()


plt.plot(data_yd,data_xd , label="icr")
plt.plot(data_yd,1/poptd[1]*data_yd -12000+1500 , label='gr={:.6f}'.format(poptd[1]))
plt.legend()
plt.show()
print(data_yd.min()*180/np.pi)
```
