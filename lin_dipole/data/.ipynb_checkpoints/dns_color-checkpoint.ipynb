```python
import numpy as np
from scipy import signal, fft
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import scipy as sp
from scipy.optimize import curve_fit 
```

```python
data=np.loadtxt("../Fld/rmhdfs_r0.1.slab_linear.run07D4_j-phi.g0000.ion.001")
```

```python
fig_set = {
    "axes.linewidth": .5,
    "axes.labelsize": 10,
    "axes.labelpad": 4.0,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8,
    "xtick.major.size": 3,
    "ytick.major.size": 3,
    "ytick.right" : True,
    "xtick.minor.visible" : False,
    "ytick.minor.visible" : True,
    "font.size": 10,
    "font.family": 'sans-serif',
    "legend.framealpha": .5, # 凡例の透明度
    "legend.frameon" : False,
    "xtick.direction": 'in',  ## デフォルト: out
    "ytick.direction": 'in'  ## デフォルト: out
}

plt.rcParams.update(fig_set)
```

```python
dt=1.0
nx=1280*2+1
nt=int(data.shape[0]/nx)
x=data[0:nx,0]
xx = 90-np.arcsin(np.sqrt(x))*180/np.pi
xc = np.arcsin(np.sqrt(x))*180/np.pi
t=dt * np.arange(nt)
dns=(data[:,1]+1j*data[:,2]).reshape(nt,nx) #密度のdataを実部＋虚部に書き直し、それを縦軸nt=1601,横軸nx=2561でreshape

dns0 =  (np.abs( dns[0,:].real ) ).max()
print(dns0)
print( (np.abs( dns[:,:].real ) ).max() / dns0)
print(np.abs(dns.real).max() )

dns = dns.real/dns0  # 初期振幅で規格化
```

```python
env = np.zeros((len(dns.real[:,0]), len(dns.real[0,:])))
# print(a.shape)

for i in range(nt):
    env[i,:] = np.abs(signal.hilbert(dns.real[i,:]))

am = np.zeros((len(dns.real[:,0])))
#xm = np.zeros((len(dns.real[:,0]), len(dns.real[0,:])))
envm=  np.zeros((len(dns.real[:,0])))
print(am.shape)
#amm = am.reshape(1,1801)
for m in range(nt):
    for n in range(nx):
        if env[m,n] > env[m,:].max() * 0.99999999999999:
            am[m] = env[m,n]
            envm[m] = xc[n]
            
plt.plot(t, envm/180*np.pi, linewidth=.8, color="b", linestyle="solid" , alpha=1.0 ,rasterized=True)   ######. 群速度
```

```python
### 2D color plot n(t,x) ###
dpi = 300
fig, ax= plt.subplots(figsize=(4,3), dpi=dpi)
ax.plot(envm,t,linewidth=.8, color="white", linestyle="solid" , alpha=1.0 ,rasterized=True)   ######. 群速度
# ax.plot(envm,t/5+600,linewidth=.8, color="white", linestyle="dashdot" , alpha=1.0 ,rasterized=True)   ######. 位相そくど
# COLATITULE
quad = ax.pcolormesh(xc,t,dns.real,rasterized=True, shading="nearest",cmap="viridis")
cbar = fig.colorbar(quad)#, shrink=1.0, aspect=11.0)
cbar.set_label('Density perturbation', fontsize=10) # 文字サイズ14でラベルを付ける)
cbar.ax.tick_params(labelsize=8,width =0.4)
cbar.outline.set_linewidth(0.4)
cbar.set_label("Numbar density")

ax.tick_params( labelsize= 8, width=0.5)
ax2 = ax.twiny()
ax2.tick_params( labelsize= 8, width=0.5)
ax2.set_xlabel("$\sin^2\\theta$" )
ax2.set_xlim(np.sin(19/180*np.pi)**2, np.sin(xc.max()/180*np.pi)**2 )
#--- COLATITUDE -----
ax.set_xlim(19, xc.max())
ax.set_ylim(0, 1700)
#--------------------
ax.set_xlabel( "Colatitude $(^\mathrm{\circ})$") 
ax.set_ylabel("Time ($\mathrm{R_I/V_A}$)")
plt.savefig("./images/fig_ion_dns.png", bbox_inches='tight', pad_inches=0.1)
plt.show()
```

```python
####### curve fit ##########
it = 300

data_xd = t
data_yd = envm/180*np.pi

parameter_initial = np.array([1 ,0.006])
def func(x, a, b ):
    fd = b*x + a
    return fd

poptd, pcovd = curve_fit(func,data_xd, data_yd, parameter_initial)

print('a*exp(b*t) -> a,b=', poptd, pcovd)

residualsd =  data_yd - func(data_xd, poptd[0],poptd[1]) #残差
rssd = np.sum(residualsd**2)  #residual sum of squares = rss
tssd = np.sum((data_yd-np.mean(data_yd))**2)  #total sum of squares = tss
r_squaredd = 1 - (rssd / tssd)

print(r_squaredd)
```

```python
plt.yscale("log")
plt.xlabel('$t~[L_x/V_A]$')
plt.ylabel('$log|J|$')
#ax = fig.add_subplot(111)
plt.plot(data_xd,data_yd , label="wave_packet")
plt.plot(data_xd,func(data_xd, poptd[0],poptd[1]) , label='gr={:.6f}'.format(poptd[1]))
plt.legend()
plt.show()


plt.plot(data_yd,data_xd , label="wave_packet")
plt.plot(data_yd, 1/poptd[1]*data_yd -12000+1500 , label='gr={:.6f}'.format(poptd[1]))
plt.legend()
plt.show()
print(data_yd.min()*180/np.pi)
```

```python
### 2D color plot n(t,x) ###
B = 1/poptd[1]

figsize_px = np.array([1280, 960])
dpi = 300
fig, ax= plt.subplots(figsize=(4,3), dpi=dpi)
ax.plot(envm,t,linewidth=.8, color="white", linestyle="solid" , alpha=1.0 ,rasterized=True)   ######. 群速度
# ax.plot(envm,t/5+600,linewidth=.8, color="white", linestyle="dashdot" , alpha=1.0 ,rasterized=True)   ######. 位相そくど
ax.plot(xc,(B*xc*np.pi/180) / 5 - 1500)

# COLATITULE
quad = ax.pcolormesh(xc,t,dns.real,rasterized=True,cmap="viridis",
                     norm=colors.SymLogNorm(linthresh=10.0, linscale=1.0, vmin=-np.abs((dns.real).min()), vmax=np.abs((dns.real).min())))
cbar = fig.colorbar(quad, shrink=1.0, aspect=10.0)
cbar.set_label('Density perturbation', fontsize=10) # 文字サイズ14でラベルを付ける)
cbar.ax.tick_params(labelsize=8,width =0.4)
cbar.outline.set_linewidth(0.4)
cbar.set_label("Numbar density")

ax.tick_params( labelsize= 8, width=0.5)
ax2 = ax.twiny()
ax2.tick_params( labelsize= 8, width=0.5)
ax2.set_xlabel("$\sin^2\\theta$" )
ax2.set_xlim(np.sin(19/180*np.pi)**2, np.sin(xc.max()/180*np.pi)**2 )
#------ COLATITUDE -------
ax.set_xlim(19, xc.max())
ax.set_ylim(0, 1700)
#-------------------------
ax.set_xlabel( "Colatitude $(^\mathrm{\circ})$") 
ax.set_ylabel("Time ($\mathrm{R_I/V_A}$)")
plt.show()
```
