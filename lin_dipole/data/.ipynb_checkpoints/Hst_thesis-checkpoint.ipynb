```python
import numpy as np
from scipy import signal, fft
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
fig_set = {
    "axes.linewidth": .5,
    "axes.labelsize": 10,
    "axes.labelpad": 4.0,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8,
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "xtick.top" : True,
    "ytick.right" : True,
    "font.size": 10,
    "font.family": 'sans-serif',
    "legend.frameon" : False,
    "xtick.direction": 'in',  ## デフォルト: out
    "ytick.direction": 'in'  ## デフォルト: out
}
plt.rcParams.update(fig_set)
```

```python
idn=np.loadtxt("./Hst/rmhdfs_r0.1.slab_linear.run07D4_j-phi.idn.001")
print(idn.shape)
t = idn[:,0]
dt=1.0
nx=1280*2+1
nt=len(t)
#t=dt * np.arange(nt)
```

```python
idn=np.loadtxt("./Hst/rmhdfs_r0.1.slab_linear.run07D4_j-phi.idn.001").reshape(nt,3) # t,total,ky0
icr=np.loadtxt("./Hst/rmhdfs_r0.1.slab_linear.run07D4_j-phi.icr.001").reshape(nt,3)
iog=np.loadtxt("./Hst/rmhdfs_r0.1.slab_linear.run07D4_j-phi.iog.001").reshape(nt,3)
mph=np.loadtxt("./Hst/rmhdfs_r0.1.slab_linear.run07D4_j-phi.mph.001").reshape(nt,3)
mps=np.loadtxt("./Hst/rmhdfs_r0.1.slab_linear.run07D4_j-phi.mps.001").reshape(nt,3)
```

```python
figsize_px = np.array([640, 480])
dpi = 250
figsize_inch = figsize_px / dpi
print("figsize", figsize_inch)  # figsize (6.4, 4.8)
fig, ax= plt.subplots(figsize=(4,3), dpi=dpi)

ax.set_yscale("log")
ax.set_xlabel('Time $\mathrm{(R_I/v_A)}$')
ax.set_xlim(0,1700)
ax.plot(t,np.sqrt(idn[:,1]),linewidth=.6 , label="$<n>$")
ax.set_ylabel("Number density")
ax.legend(fontsize=7).get_frame().set_linewidth(.7)
plt.show()

figsize_px = np.array([640, 480])
dpi = 250
figsize_inch = figsize_px / dpi
print("figsize", figsize_inch)  # figsize (6.4, 4.8)
fig, ax= plt.subplots(figsize=(4,3), dpi=dpi)
ax.set_yscale("log")
ax.set_xlabel('Time $\mathrm{(R_I/v_A)}$')
ax.set_ylabel('Current density and vorticity')
ax.set_xlim(0,1700)
ax.plot(t,np.sqrt(icr[:,1]),linewidth=.6 ,linestyle='solid', label="$<J>$")
ax.plot(t,np.sqrt(iog[:,1]) ,linewidth=.6 , linestyle="dashed", label="$<\omega>$")
ax.legend(fontsize=7).get_frame().set_linewidth(.7)
plt.show()


figsize_px = np.array([640, 480])
dpi = 250
figsize_inch = figsize_px / dpi
print("figsize", figsize_inch)  # figsize (6.4, 4.8)
fig, ax= plt.subplots(figsize=(4,3), dpi=dpi)
ax.set_yscale("log")
ax.set_xlabel('Time $\mathrm{(R_I/v_A)}$')
ax.set_ylabel('Stream function and \n parallel vector potential')
ax.set_xlim(0,1700)
ax.plot(t,np.sqrt(mph[:,1]),linewidth=.6 ,linestyle='solid', label="$<\phi>$")
ax.plot(t,np.sqrt(mps[:,1]) ,linewidth=.6 , linestyle="dashed", label="$<\psi>$")
ax.legend(fontsize=7).get_frame().set_linewidth(.7)
plt.show()
```
