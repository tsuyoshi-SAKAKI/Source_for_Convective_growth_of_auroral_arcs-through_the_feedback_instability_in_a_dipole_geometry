```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
data = np.loadtxt("./Fld/rmhdfs_r0.1.slab_linear.run07D4_j-phi.g0000.ion.001")
```

```python
fig_set = {
    "axes.linewidth": .5,
    "axes.labelsize": 10,
    "axes.labelpad": 4.0,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8,
    "xtick.major.size": 3,
    "ytick.major.size": 3,
    "ytick.right" : True,
    "xtick.minor.visible" : False,
    "ytick.minor.visible" : True,
    "font.size": 10,
    "font.family": 'sans-serif',
    "legend.framealpha": .5, # 凡例の透明度
    "legend.frameon" : False,
    "xtick.direction": 'in',  ## デフォルト: out
    "ytick.direction": 'in'  ## デフォルト: out
}
plt.rcParams.update(fig_set)
```

```python
dt=1.0
nx=1280*2+1
nt=int(data.shape[0]/nx)
x=data[0:nx,0]
xx = 90-np.arcsin(np.sqrt(x))*180/np.pi
xc = np.arcsin(np.sqrt(x))*180/np.pi
t=dt * np.arange(nt)

#密度のdataを実部＋虚部に書き直し、それを縦軸nt=1601,横軸nx=2561でreshape, 初期振幅で規格化
dns= (data[:,1]+1j*data[:,2]).reshape(nt,nx)
dns= dns / (np.abs( dns[0,:].real ) ).max()
omg= (data[:,3]+1j*data[:,4]).reshape(nt,nx)
omg= omg / (np.abs( omg[0,:].real ) ).max()
cpr= (data[:,5]+1j*data[:,6]).reshape(nt,nx)
cpr= cpr / (np.abs( cpr[0,:].real ) ).max()
```

```python
a = np.zeros((len(dns.real[:,0]), len(dns.real[0,:])))
print(a.shape)

for i in range(nt):
    a[i,:] = np.abs(signal.hilbert(dns.real[i,:]))

am = np.zeros((len(dns.real[:,0])))
xm=  np.zeros((len(dns.real[:,0])))
print(am.shape)
for m in range(nt):
    for n in range(nx):
        if a[m,n] > a[m,:].max() * 0.99999999999999:
            am[m] = a[m,n]
            xm[m] = xc[n]
```

```python
### 2D color plot n(t,x) ###
dpi = 300
fig, ax= plt.subplots(figsize=(4,3), dpi=dpi)

#ax.plot(xm,t,linewidth=.8, color="white", linestyle="solid" , alpha=1.0 ,rasterized=True)   ######. 群速度
# COLATITULE
quad = ax.pcolormesh(xc,t,dns.real,rasterized=True,cmap="viridis")
cbar = fig.colorbar(quad)
cbar.ax.tick_params(labelsize=8,width =0.4)
cbar.outline.set_linewidth(0.4)
cbar.set_label("Numbar density")
ax.tick_params( labelsize= 8, width=0.5)
ax2 = ax.twiny()
ax2.tick_params( labelsize= 8, width=0.5)
ax2.set_xlabel("$\sin^2\\theta$" )
ax2.set_xlim(np.sin(19/180*np.pi)**2, np.sin(xc.max()/180*np.pi)**2 )
# COLATITUDE
ax.set_xlim(19, xc.max())
ax.set_ylim(0, 1700)
# COLATITUDE
# ax.set_xlabel( "Colatitude $(^\mathrm{\circ})$") 
ax.set_ylabel("Time ($\mathrm{R_I/V_A}$)")
ax.text(17.8, 1850, "a", fontfamily="sans-serif", fontweight="bold", fontsize=12)
plt.savefig("./Fig/fig_ion_dns.png", bbox_inches='tight', pad_inches=0.1)
plt.show()
```
